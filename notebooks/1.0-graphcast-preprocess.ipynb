```python
lon_space = 2 #graphcast.lon.diff('lon').mean()
lat_space = 2 #graphcast.lat.diff('lat').mean()

lons = np.arange(aorc.longitude.min(),aorc.longitude.max(),lon_space)
lats = np.arange(aorc.latitude.min(),aorc.latitude.max(),lat_space)

graphcast_norm_lon = graphcast.copy()
graphcast_norm_lon = graphcast_norm_lon.assign_coords({'lon':sorted([lon if lon<180 else lon-360 for lon in graphcast_norm_lon.lon.values])}).rename({'lat':'latitude','lon':'longitude'})
graphcast_norm_lon = graphcast_norm_lon.interp(latitude=lons, longitude=lats,method='linear')
display(graphcast_norm_lon)

aorc_round = aorc.sel(time=slice(graphcast_norm_lon.time.min(),graphcast_norm_lon.time.max())).resample(time='1D').sum().interp(latitude=lons, longitude=lats,method='linear')
aorc_round
```
